```python
# Import pandas library for working with tables (dataframes)
# Used heavily in data analysis and telecom KPI datasets
import pandas as pd

# Import numpy library for numerical operations
# Useful for calculations, arrays, and mathematical operations
import numpy as np

# Import matplotlib for creating graphs and visualizations
# Used to plot charts like bar charts, scatter plots, etc.
import matplotlib.pyplot as plt

# Enable inline plotting in Jupyter Notebook
# This ensures graphs are displayed inside the notebook itself
%matplotlib inline
```

```python
# Create a dictionary that simulates telecom KPI data
# Each key represents a KPI metric in a telecom network

data = {
    # Cell IDs represent different network towers or base stations
    "cell_id": ["A101", "A102", "A103", "A104", "A105"],

    # Throughput = speed of data transfer in Mbps
    "throughput_mbps": [45, 78, 32, 90, 55],

    # Latency = delay in milliseconds (lower is better)
    "latency_ms": [25, 40, 55, 20, 35],

    # Packet loss percentage = data lost during transmission
    "packet_loss_%": [0.5, 1.2, 3.0, 0.3, 0.8],

    # Number of active users connected to each cell
    "users_connected": [120, 200, 150, 300, 180]
}

# Convert the dictionary into a pandas DataFrame (table format)
# This makes it easy to analyze and manipulate data
df = pd.DataFrame(data)

# Display the DataFrame so we can visually inspect the dataset
df
```

```python
# .describe() gives statistical summary of numeric columns
# It shows mean, min, max, standard deviation, etc.
# This helps understand overall network performance
df.describe()
```

```python
# Filter rows where packet loss is greater than 1%
# These represent problematic or weak network cells
poor_cells = df[df["packet_loss_%"] > 1.0]

# Display only the poor-performing cells
# This helps identify network issues quickly
poor_cells
```

```python
# Create a scatter plot to analyze relationship between latency and throughput

# X-axis = latency (delay in network)
# Y-axis = throughput (speed of network)

plt.scatter(df["latency_ms"], df["throughput_mbps"])

# Add title to the chart for clarity
plt.title("Latency vs Throughput Analysis")

# Label X-axis to show what it represents
plt.xlabel("Latency (ms)")

# Label Y-axis to show throughput values
plt.ylabel("Throughput (Mbps)")

# Display the graph on screen
plt.show()
```

```python
# Create a bar chart to show packet loss per cell

# X-axis = cell IDs (A101, A102, etc.)
# Y-axis = packet loss percentage

plt.bar(df["cell_id"], df["packet_loss_%"])

# Add chart title
plt.title("Packet Loss per Network Cell")

# Label X-axis
plt.xlabel("Cell ID")

# Label Y-axis
plt.ylabel("Packet Loss (%)")

# Display the bar chart
plt.show()
```

```python
# Define a function to calculate a simple KPI score for each cell
# This simulates a basic AI/ML scoring logic used in telecom optimization

def kpi_score(row):
    # Start with a perfect score of 100
    score = 100

    # Penalize score based on latency (higher delay = worse performance)
    score = score - (row["latency_ms"] * 0.5)

    # Penalize score based on packet loss (very critical in telecom)
    score = score - (row["packet_loss_%"] * 10)

    # Return final calculated score rounded to 2 decimal places
    return round(score, 2)

# Apply the function to each row of the DataFrame
# This creates a new column called "kpi_score"
df["kpi_score"] = df.apply(kpi_score, axis=1)

# Display updated DataFrame with KPI scores
df
```

```python
# Find the row with highest KPI score (best network performance)
best_cell = df.loc[df["kpi_score"].idxmax()]

# Find the row with lowest KPI score (worst network performance)
worst_cell = df.loc[df["kpi_score"].idxmin()]

# Print results clearly for analysis
print("🏆 Best Performing Cell:")
print(best_cell)

print("\n⚠️ Worst Performing Cell:")
print(worst_cell)
```
